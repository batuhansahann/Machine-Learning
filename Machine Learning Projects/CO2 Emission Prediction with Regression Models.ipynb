# <font color='green'> <b>Simple Linear Regression Model</b><font color='black'>

**Lineer Regresyon**

Bamsz deikenler zerinden baml deikeni tahmin eder.

Tek bamsz deiken varsa simple, birden ok bamsz deiken varsa multiple lineer regresyon ad verilir.

Simple linear regression'da feature ile target arasnda anlaml bir iliki varsa bu data linear regression iin uygundur


**Lineer regression analizinin temel varsaymlar**

1. **Lineer liki Varsaym:** Baml deiken ile bamsz deikenler arasndaki ilikinin lineer bir ekilde ifade edilebilecei varsaylr. Yani, regresyon dorusu ile deikenler arasndaki iliki lineer olmaldr. rnein, X arttka Y de artmal veya azalmaldr.

2. **Bamszlk Varsaym:** Gzlemler arasnda bamszlk varsaylr. Yani, bir gzlemin sonucu dier gzlemleri etkilememelidir.

3. **Normal Dalm Varsaym:** Hata terimleri (rezideller) normal dalml olmaldr ve herhangi bir pattern oluturmamaldr. Bu, regresyon modelinin tahminlerinin gvenilir olmas iin gereklidir.

4. **Bamsz Deikenlerin Bamszl:** Bamsz deikenler arasnda oklu dorusallk (multicollinearity) sorunu olmamaldr. Yani, bamsz deikenler birbirine ok yakn ve yksek derecede korele olmamaldr.

- Yapka = b0 + b1X

- Yapka= tahmin edilen deer

- b0= intercept (dorunun y eksenini kestii nokta)

- b1= slope = coefficient = weight

- X=bamsz deiken

- Residual = Random error = e= Y-Yapka

**nemli olan erroru, hatay minimize etmektir.**

**Best fit line**; hatalarmz minimize edecek ekilde iziyor. Tek bir feature olduunda **Ordinary Least Squares** metot ile, birden fazla olduunda ise **gradient descent** ile bulunur.

**Cost - loss function**; gerek deer ile tahmin edilen deerlerlerin farknn karesini alp ortalama hatay buluyor.

**Gradient descent** optimizasyon algoritmasyla, yapacanz ey, w ve b deerini yava yava deitirerek cost - lost fonksiyonunu azaltmak ve minimum deerine getirmeye almak olacak.

teratif ekilde hatay minimize etmeye alan bir algoritmadr.

**Bias**, bir modelin tahminlerindeki sistematik hatadr. bir modelin tahminleri gerek deerlerden ne kadar uzaksa, o kadar yksek bir bias vardr. Yksek bias, modelin basit olmas ve verilerin karmakln tam olarak yakalayamamas nedeniyle ortaya kabilir. Underfitting dolaysyla

**Variance** ise, modelin farkl veri noktalar zerindeki tahminlerinin ne kadar deiken olduunu ifade eder. 
Ayn veri kmesi zerinde birden ok kez eitildiinde farkl sonular elde edilir.
Overfitting sonucunda ortaya kar.
Yksek bir varyans, modelin verilere ar uyum saladn ve rendii rntlerin veri seti dndaki yeni verilere genelletirilemeyeceini gsterir.

**bias-varyans trade-off**:
Bias ve variance arasnda bir denge kurmak nemlidir. deal olarak, bir modelin hem dk bias hem de dk variance deerine sahip olmas gereklidir. Bu, modelin verilerin karmakln doru bir ekilde yakalayabilmesi ve genelleme yapabilmesi anlamna gelir.




## <font color='Orange'> <b>Splitting the dataset into X (independent variables) and y (dependent variable)</b><font color='black'>

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_slr = df[['Engine_Size_L']]
y_slr = df['CO2_Emissions_g_per_km']
```

## <font color='blorange'> <b>Train | Test Split</b><font color='black'>

```python
X_train, X_test, y_train, y_test = train_test_split(X_slr, y_slr, test_size=0.2, random_state=42)
```

## <font color='orange'> <b>Training the Model</b><font color='black'>

```python
model_slr = LinearRegression()
model_slr.fit(X_train, y_train)
```

## <font color='orange'> <b>Predicting Test Data</b><font color='black'>

```python
y_pred = model_slr.predict(X_test)
```

## <font color='orange'> <b>Evaluating the Model</b><font color='black'>

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R2: {r2}")
```

## <font color='Orange'> <b>Control of mistake values</b><font color='black'>

```python
residuals = y_test - y_pred
sns.histplot(residuals, kde=True)
plt.show()
```

# <font color='green'> <b>Multiple Linear Regression Model</b><font color='black'>

## <font color='orange'> <b>Splitting the dataset into X(independent variables) and y (dependent variable)</b><font color='black'>

```python
X_multi = df[['Engine_Size_L', 'Fuel_Consumption_Comb_L_per_100_km', 
            'Fuel_Consumption_Hwy_L_per_100_km', 'Fuel_Consumption_City_L_per_100_km']]
Y_multi = df['CO2_Emissions_g_per_km']
```

## <font color='orange'> <b>Training the Model</b><font color='black'>

```python
model_multi = LinearRegression()
model_multi.fit(X_train_multi, y_train_multi)
```

## <font color='orange'> <b>Predicting Test Data</b><font color='black'>

```python
y_pred_multi = model_multi.predict(X_test_multi)
```

## <font color='orange'> <b>Evaluating the Model</b><font color='black'>

```python
mse_m = mean_squared_error(y_test_multi, y_pred_multi)
r2_m = r2_score(y_test_multi, y_pred_multi)
```

```python
print(f"Mean Squared Error (MSE): {mse_m}")
print(f"R-squared: {r2_m}")
```

```python
print("Model Coef:", model_multi.coef_)
print("Intercept:", model_multi.intercept_)
```
